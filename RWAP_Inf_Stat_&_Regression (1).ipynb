```python
# Create a new DataFrame as a copy of the original
df2 = df1.copy()

# Function to split columns into multiple columns
def split_column(df, column_name):
    if column_name in df.columns:
        split_df = df[column_name].str.split(', ', expand=True)
        split_df.columns = [f'{column_name}_{i+1}' for i in range(split_df.shape[1])]
        return pd.concat([df, split_df], axis=1).drop(columns=[column_name])


# Apply the function to 'player_positions', 'player_tags', and 'player_traits'
df2= split_column(df2, 'player_positions')
df2= split_column(df2, 'player_tags')
df2 = split_column(df2, 'player_traits')

# Check the updated DataFrame
df2.head(5)
```

```python
# Function to remove special characters and handle NaN values
def clean_text(text):
    if pd.isna(text):  # Check if the value is NaN
        return text
    return unidecode(text)

# Apply the cleaning function to the columns
df2['short_name'] = df2['short_name'].apply(clean_text)
df2['club_name'] = df2['club_name'].apply(clean_text)
df2['league_name'] = df2['league_name'].apply(clean_text)
df2['nationality_name'] = df2['nationality_name'].apply(clean_text)

# Display the cleaned data
df2.head()
```

```python
# Function to sum values in the format 'number+number'
def sum_values(value):
    if pd.isna(value):  # Check for NaN values
        return value
    # Convert value to string to handle both integers and strings
    value = str(value)
    # Split the string by '+', convert to integers, and sum
    parts = value.split('+')
    if len(parts) == 2:  # Ensure there are exactly two parts
        return int(parts[0]) + int(parts[1])
    return value

# Apply the function to each column that needs summing
columns_to_sum = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']

for column in columns_to_sum:
    df2[f'{column}_sum'] = df2[column].apply(sum_values)

# Display the updated DataFrame
df2.head(5)
```

**INFERENTIAL STATISTICS**

```python
from scipy.stats import f_oneway
from scipy.stats import pearsonr
import statsmodels.api as sm
```

### To check differences between wages for top 3 countries:

```python
# Assuming df2 is your DataFrame
# Filter data for the three countries
armenia_wages = df2[df2['nationality_name'] == 'Armenia']['wage_eur']
suriname_wages = df2[df2['nationality_name'] == 'Suriname']['wage_eur']
burkina_faso_wages = df2[df2['nationality_name'] == 'Burkina Faso']['wage_eur']

# Perform ANOVA
f_stat, p_value = f_oneway(armenia_wages, suriname_wages, burkina_faso_wages)

print(f"F-Statistic: {f_stat}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in average wages among players from Armenia, Suriname, and Burkina Faso.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average wages among players from Armenia, Suriname, and Burkina Faso.")
```

### correlation between ranking and market value

```python
# Assuming df2 is your DataFrame
# Drop rows with missing values in 'overall' or 'value_eur'
data_clean = df2[['overall', 'value_eur']].dropna()

# Compute Pearson correlation
corr, p_value = pearsonr(data_clean['overall'], data_clean['value_eur'])

print(f"Correlation Coefficient: {corr}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant correlation between player rankings and market value.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between player rankings and market value.")
```

**Regression**


### Null Hypothesis (H₀): The independent variables (overall, potential, age, wage_eur, international_reputation, player_positions, skill_moves) have no significant effect on the dependent variable (value_eur)
### Alternative Hypothesis (H₁): At least one of the independent variables has a significant effect on the dependent variable (value_eur)

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming df2 is your DataFrame
# Drop rows with missing values in relevant columns
df_clean = df2[['value_eur', 'overall', 'potential', 'age', 'wage_eur', 'international_reputation', 'player_positions_1', 'skill_moves']].dropna()

# Convert categorical variable to one-hot encoded variables
df_clean = pd.get_dummies(df_clean, columns=['player_positions_1'], drop_first=True)

# Define the regression model
model = smf.ols('value_eur ~ overall + potential + age + wage_eur + international_reputation + skill_moves + ' + ' + '.join(df_clean.columns[df_clean.columns.str.startswith('player_positions_1')]), data=df_clean).fit()

# Print the summary
print(model.summary())
```

### Null Hypothesis (H₀): The independent variables (age, potential, value_eur, wage_eur, preferred_foot, skill_moves, work_rate, height_cm, weight_kg) have no significant effect on the overall rating.
### Alternative Hypothesis (H₁): At least one of the independent variables has a significant effect on the overall rating.

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming df2 is your DataFrame
# Drop rows with missing values in relevant columns
df_clean = df2[['overall', 'age', 'potential', 'value_eur', 'wage_eur', 'preferred_foot', 'skill_moves', 'work_rate', 'height_cm', 'weight_kg']].dropna()

# Convert categorical variables to one-hot encoded variables
df_clean = pd.get_dummies(df_clean, columns=['preferred_foot', 'work_rate'], drop_first=True)

# Define the regression model
model = smf.ols(
    'overall ~ age + potential + value_eur + wage_eur + skill_moves + height_cm + weight_kg + ' +
    ' + '.join(df_clean.columns[df_clean.columns.str.startswith('preferred_foot_')]) ,
    data=df_clean).fit()

# Print the summary
print(model.summary())
```
