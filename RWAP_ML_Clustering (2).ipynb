```python
# Function to remove special characters and handle NaN values
def clean_text(text):
    if pd.isna(text):  # Check if the value is NaN
        return text
    return unidecode(text)

# Apply the cleaning function to the columns
df2['short_name'] = df2['short_name'].apply(clean_text)
df2['club_name'] = df2['club_name'].apply(clean_text)
df2['league_name'] = df2['league_name'].apply(clean_text)
df2['nationality_name'] = df2['nationality_name'].apply(clean_text)

# Display the cleaned data
df2.head()
```

```python
# Function to sum values in the format 'number+number'
def sum_values(value):
    if pd.isna(value):  # Check for NaN values
        return value
    # Convert value to string to handle both integers and strings
    value = str(value)
    # Split the string by '+', convert to integers, and sum
    parts = value.split('+')
    if len(parts) == 2:  # Ensure there are exactly two parts
        return int(parts[0]) + int(parts[1])
    return value

# Apply the function to each column that needs summing
columns_to_sum = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']

for column in columns_to_sum:
    df2[f'{column}_sum'] = df2[column].apply(sum_values)

# Display the updated DataFrame
df2.head(5)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score as dbscore, silhouette_score as sscore
import matplotlib.pyplot as plt

# Assuming combined_data is already preprocessed and numeric
# Dropping the 'index' column if it's included in combined_data
data_for_clustering = combined_data.drop(columns=['index'])
```

```python
import numpy as np
import pandas as pd

# Assuming df3 is your DataFrame after dropping certain columns
numeric_df3 = data_for_clustering.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df3.corr().abs()

# Select the upper triangle of the correlation matrix
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(np.bool_))

# Find index of feature columns with correlation greater than 0.8
high_corr_pairs = [(col, correlation_matrix.columns[i]) for i in range(len(upper_triangle.columns)) for col in upper_triangle.columns if upper_triangle[col][i] > 0.8]

print("Highly Correlated Features (correlation > 0.8):")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]} with correlation {correlation_matrix.loc[pair[0], pair[1]]}")

# Create a set to keep track of columns to drop
cols_to_drop = set()

# Add the first feature in each high correlation pair to the drop list
for pair in high_corr_pairs:
    cols_to_drop.add(pair[0])

print("\nColumns to consider dropping:")
for col in cols_to_drop:
    print(col)
```

```python
# Dropping the columns from df3
df4 = data_for_clustering.drop(columns=cols_to_drop)
```

```python


# Initialize lists to store the scores
db_scores = []
ss_scores = []

# Range of cluster numbers to evaluate
cluster_range = range(2, 11)

for k in cluster_range:
    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(df4)

    # Calculate Davies-Bouldin score
    db_score = dbscore(df4, cluster_labels)
    db_scores.append(db_score)

    # Calculate Silhouette score
    ss_score = sscore(df4, cluster_labels)
    ss_scores.append(ss_score)

# Plotting the scores
plt.figure(figsize=(12, 6))

# Davies-Bouldin Score plot
plt.subplot(1, 2, 1)
plt.plot(cluster_range, db_scores, marker='o', color='r')
plt.title('Davies-Bouldin Score')
plt.xlabel('Number of Clusters')
plt.ylabel('DB Score')

# Silhouette Score plot
plt.subplot(1, 2, 2)
plt.plot(cluster_range, ss_scores, marker='o', color='b')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('SS Score')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming your clusters are segregated into different DataFrames
clusters = [football_cluster0, football_cluster1, football_cluster2, football_cluster3]

# Dictionary to store dominant variables for each cluster
dominant_variables = {}

for i, cluster in enumerate(clusters):
    # Calculate the mean of each variable in the cluster
    means = cluster.mean()

    # Identify the variable with the highest mean value
    dominant_variable = means.idxmax()

    # Store the dominant variable for this cluster
    dominant_variables[f'Cluster {i}'] = dominant_variable

# Print the dominant variable for each cluster
for cluster, variable in dominant_variables.items():
    print(f"{cluster}: Dominant Variable -> {variable}")
```

```python
# Define a function to get the dominant category for each cluster
def get_dominant_category(cluster_data):
    return cluster_data.idxmax()

# Create a dictionary to store the dominant categories for each cluster and variable
cluster_profiles = {}

relevant_Columns=football_cluster.columns
# Iterate over relevant columns
for column in relevant_Columns:
    # Group the data by 'Cluster_Label' and get value counts for each cluster
    cluster_freq = football_cluster.groupby('cluster_number')[column].value_counts()

    # Find the dominant category for each cluster
    dominant_categories = cluster_freq.groupby(level=0).apply(get_dominant_category)

    # Store the dominant categories in the dictionary
    cluster_profiles[column] = dominant_categories

# Print the cluster profiling
for column, dominant_categories in cluster_profiles.items():
    print(f"Variable: {column}")
    for cluster, category in dominant_categories.items():
        print(f"Cluster {cluster}: Dominant category - {category}")
    print()
```
